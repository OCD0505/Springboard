**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#creatinbg the function as described above

def create_price_ratio (x,y):
    
    # Passed the dataframe to x and the London Borough name to y 
    # Subset x with y, used a query to look at the specific row and indexed x to look at the specific column
    # Finally, I used iloc to pull the value at the intersection of the selected 'Year' and 'Avg Price'
    # I stored the values in variables to use in the ratio calculation 
    price_1998 = x[x['London Borough']==y].query('Year == 1998')['Avg Price'].iloc[0]
    price_2018 = x[x['London Borough']==y].query('Year == 2018')['Avg Price'].iloc[0]
    
    return price_1998/price_2018

# Initialized max_ratio to have the first comparison value for the if conditional below
max_ratio=0
# Delared dictionary to store London Boroughs and their correspond 1998/2018 ratio 
ratio_dict={}

for i in df_boroughs_r1['London Borough'].unique():
    # storing each ratio in the dictionary 
    ratio_dict[i] = create_price_ratio(df_boroughs_r1,i)
    # checking each value to find the max
    # If the current dictionary value is greater than the previously determined max, then it will
    #replace the previous value in max_ratio
    if ratio_dict[i]>max_ratio: 
        max_ratio = ratio_dict[i]
        
# Used diciotnary created in loop above to create a data frame of Boroughs and Ratios
# Added sort method to get values sorted from greatest to least
ratio_df = pd.DataFrame(ratio_dict.items(), columns=['London Borough','Ratio']).sort_values('Ratio',ascending=False).reset_index(drop=True)

# Sliced the dataframe above and stored the top 15 Boroughs & Ratios in a variable, top15
top15 = ratio_df.iloc[0:15,:]

# Used top15 to create a bar graph to show which Borough had the greatest ratio 
#(ie greatest price increase between 1998 - 2018)
ax=top15.plot(kind='bar')
ax.set_xticklabels(top15['London Borough'])

plt.show()
```
